# Dynamic Arrays

## Concept

Dynamic arrays are essentially an ordered table of datatypes (integers, strings, floats, objects, etc). Dynamic arrays have the added benefit of being resizable (ex. when inserting, deleting, splicing, appending, etc) where as in other cases a new array would have to be created if the user wanted to change the size of the array. Dynamic arrays are useful for grouping a series of datatypes of varying length/number. If the number of elements in an array is fixed, use a standard array to be more efficient in time/space complexity. Elements in the array are orded by index and in memory are stored next to eachother based on the starting address, index, and data size (address = base_address + index*byte_size). Dynamic arrays can either be homogeneous or heterogeneous meaning they will have all the same data types or can have multiple (ex, lists in python are heterogenous dynamic arrays).

#Searching Algorithms

##Linear Search

*   ***Description:*** Linear search works by looping through a data set and comparing the value your looking at and the value you are trying to find. When there is a match it returns the location. Can be used on either sorted or unsorted datasets.
*   ***Time complexity:*** The best case would be that the data is the first one the program looks at and hence the program would only run once maling it a constant time complexity O(1). The average case would be in the middle of the dataset where the program would run n number of times making the algorithm time complexity linear O(n). The worst case would be at the end of the dataset which the algorith would run the length of the dataset which would also be linear O(n). It is also important to note that this is assuming that the data set itself is linear and one-dimensional. For higher dimensional datasets you would increase the average and worst case time complexity by a factor of m since you would need to loop through m number of subsets O(n*m).

ref: https://www.geeksforgeeks.org/searching-algorithms-for-a-2d-arrays-matrix/ 

```
# standard binary search for array
def linear_search(val, arr):
  # loop through dataset at each index
  for i in range(len(arr)):
    # return index if value is found
    if val == arr[i]:
      return i  
  # return error message if value was not found in dataset
  return print('Value not found.')

# 2D array binary search example
def linear_search_2d(val, arr_2d):
  for i in range(len(arr_2d)):
    for j in range(len(arr_2d[i])):
      if arr_2d[i][j] == val:
        return [i, j]
  return print('Value not found.')
```

##Binary Search

 
*   ***Description:*** Binary search algorithms only for sorted datasets. It works by finding the middle of a set and then checking if the value is 'greater than' or 'less than' the middle value and then 'removes' the half that doesn't contain the target value. Then it recurses through the function using the new half until the middle value is the target value and then returns its location in the overall dataset.
*   ***Time complexity:*** Best case would be that the target value is the exact middle so the program would only run once therefore time complexity is O(1). The average case (not middle but not last recursion) and worst case (last recursion) would both be O(logn). First iteration length = n, second length = n/2, third length = n/2^2, and therefore kth iteration length = n/2^k. The length of array is approaching 1 since worst case there would be one value left to check. So n/2^k = 1 which equates to log_2(n) = k. Therefore for k iterations the time complexity would be O(logn)

ref: https://www.geeksforgeeks.org/complexity-analysis-of-binary-search/

```
# binary search using recursion
def binary_search(arr, hi, lo, val):
  # check that length of arr is greater than one
  if hi >= lo:
    # find middle index
    mid = (hi - low) // 2
    # check right side
    if arr[mid] < val:
      return binary_search(arr, mid + 1, hi, val)
    # check left side
    elif arr[mid > val]:
      return binary_search(arr, lo, mid - 1, val)
    # otherwise middle value = target value, so return middle index
    else:
      return mid
  # otherwise target value is not in dataset
  else:
    return print('Value not found.')
```

# Sorting Algoritms

##Inserion Sort

* ***Description:*** More simplistic sorting algorithm. It essentially works similar to how players would sort cards in their hand. You start at the front, move the card to the right position only moving backwards (in a realistic scenario you could move it forward too and move groupings of cards but for the sake of the analogy we will keep it simple), then move to the next card and repeats over and over until the whole deck is sorted.
* ***Time complexity:*** Best case (already sorted) would be O(n) since you still have to iterate through the entire dataset. Average (some sorting) and worst case (sort every value) would be O(n^2) since first you iterate through each index then at each index you iterate n number of times until the value is at the right index.
* ***Space complexity:*** Would be O(1) for any case since no new dataset is needed to be created to sort the given dataset.
* ***Benefits & Drawbacks:*** One benefit would be the space required to run but the drawback is the time complexity. Therefore, this would work best with small, simple datasets.

```
# example of insertion sort
def insertion_sort(arr):
  
  for i in range(1, len(arr)):      # iterate through each value
    key = arr[i]                    # look at next element
    j = i-1                         # and compare to previous
    while j >= 0 and key < arr[j]:  # move all elments greater than key in front
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key
```

##Selection Sort

* ***Description:*** This algorithm looks at the whole array, finds the smallest value, moves it to the front, then repeats the process continuelly looking at one less values in the array. In other words it takes the smalles value in an unsorted section and moves it to the front which would be then considered the sorted section so the sorted section grows while the unsorted shrinks to zero.
* ***Time complexity:*** In any case it would be O(n^2) since first it loops through each element then loops throught the unsorted section to move the minimum value. 
* ***Space complexity:*** The space the algorithm takes up remains constant O(1) since the algorithm doesn't create any new data structures to sort the original dataset. It only makes changes to the dataset. 
* ***Benefits & Drawbacks:*** Like with the insertion sort, this algorithm would work best with smaller datasets since as the dataset increases so does the time it takes to complete the sort.

```
# example of selection sort
# ref: https://www.geeksforgeeks.org/selection-sort/ 
def selection_sort(arr):
  
  for i in range(len(arr)):           # iterate through each element in array
    min_i = i                         # set minimum value index, default to current element
    for j in range(i+1, len(arr)):    # find the actual minimum index by comparing elements
      if arr[min_i] > arr[j]:         # set min index to lower value in comparison
        min_i = j
    # swap min value with first element in unsorted set
    arr[i], arr[min_i] = arr[min_i], arr[i]
```

##Merge Sort

* ***Description:*** Works through a divide and conquer method. it divides the dataset into subsets that are then sorted and then merged together into one sorted dataset.
* ***Time complexity:*** The time complexity is O(n logn) for all cases. Similar to binary search, the algorithm splits the dataset in half recursively, however it additionally iterated through the subsets to sort them hence O(n) * O(logn) = O(nlogn)
* ***Space complexity:*** The space complexity is O(n) since it has to create n number of subsets from the data set to sort/merge.
* ***Benefits & Drawbacks:*** One benefit is the time complexity as it is faster than the previous sorts however it takes more space. Therefore it would be more efficient sorting larger data sets that would otherwise take longer to sort using other algorithms.

ref: https://www.geeksforgeeks.org/merge-sort/ 

```
# example of merge sort
# source: https://www.geeksforgeeks.org/merge-sort/ 
def merge_sort(arr):
  if len(arr) > 1:
         # Finding the mid of the array
        mid = len(arr)//2
 
        # Dividing the array elements
        L = arr[:mid]
        # into 2 halves
        R = arr[mid:]
 
        # Sorting the first half
        merge_sort(L)
        # Sorting the second half
        merge_sort(R)
 
        i = j = k = 0
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
 
```

##Quick Sort

* ***Description:*** This algorithm is also a divide and conquer method similar to merge sort. It first chooses an element to act as a pivot point and divides the data set off of that pivot point. This pivot point location is arbitrary. It can be anywhere on the dataset: first, last, median, or random. After a value is selected it "partitions" the dataset based on what values are less than or greater than the pivot element.
* ***Time complexity:*** 
* ***Space complexity:***

```
# example of quick sort using recursion
# ref: https://www.geeksforgeeks.org/quick-sort/ 
def quick_sort(arr, lo, hi):
  piv_i = partition(arr, lo, hi)  # find pivot point
  quick_sort(arr, lo, piv_i - 1)  # recurse left of pivot point
  quick_sort(arr, piv_i + 1, hi)  # and recurse right 

def partition(arr, lo, hi):
  pivot = arr                     # choose right most (last) element as pivot point
  i = lo - 1                      # set pointer
  for j in range(lo, hi):         # iterate through range of values
    if arr[j] <= pivot:           # if element is less than or equal to pivot then
      i += 1                      # swap it with greater element using pointer
      arr[i], arr[j] = arr[j], arr[i]
```

## Concept

A hash table (aka hash map) is an abstract data type that maps 'keys' to 'values' similar to words in a dictionary. It uses a hash function to compute an index from a given key to an assigned bucket where the key-value pair will be stored. Ideally you want one bucket per k-v pair, however, that doesn't always occur. Sometimes different keys have the same hashed index, this is called a collision. There are multiple ways of handling collisions: chaining, rehashing, etc. Chaining occurs when you append a key-value pairs to the same index essentially creating a 3D array (which increases the time complexity of most operations). Rehashing occurs when the collision is detected and the key is rehashed to a different index that is (hopefully) empty. 

ref: https://en.wikipedia.org/wiki/Hash_table

## API & Efficiency

***Description:*** Hash tables are usually implemented using dynamic array data structures, which makes updating, inserting, deleting, etc. more efficient.

*   _hash:_ Hashes the key to derive an index for a key-value pair using some hashing algorithm. The time complexity deoends on what hashing algorith that is chosen. the desired time complexity is desired to be as low as possible O(1) so one algorith is run only once per key given.
*   _set_item/ get_item:_ Sets/gets the key value pair in one of the buckets of the hash table. This is a O(n) when chaining occurs as it has to loop through to find the right key in one of the buckets. Otherwise, it is O(1)
*   _delete:_ Used to delete a key-value pair based on a given key from the hash table. If each pair has a unique bucket the time complexity is O(1), one constant algorithm per piece of data in the hash table.

The following code implements my set method from my hash table implementation.  This code's time complexity is O(n) because it assumes chaining occurs and traverses the bucket to update values.


```

# ref: https://www.analyticsvidhya.com/blog/2021/08/explore-the-magic-methods-in-python/ 
# ref: https://coderbook.com/@marcus/how-to-create-a-hash-table-from-scratch-in-python/ 
def __setitem__(self, key, val):
    index = self.hash(key)
    if self.values[index] is not []:
        for kvp in self.values[index]:
            if kvp[0] == key:    # update existing key
                kvp[1] = val
                break
        else:                    # add new k-v pair to existing index
            self.values[index].append([key, val])
            self.size += 1
    else:                        # add new k-v pair to empty index
        self.values[index] = [[key, val]]
        self.size += 1
```

## Concept

A linked list is essentially a series of linked lists each containing one value and pointers that point to the address of the next/previous node. These one-value lists are known as nodes. There are multiple types of linked lists: singly, doubly, circular, and doubly-circular. Singly linked lists traversal only occurs in one direction (nodes only contain a next pointer). Doubly linked lists traversal can occur in both directions (nodes use next and previous pointers). In circular linked lists once the end of the list is reached it points back to the beginning (usually a sentinal node containing no data). In doubly circular linked lists the sentinal node can be pointed to or at the last node (traversal is the most flexible).

# Stack

## Concept

A stack is a data structure that is essentially how it sounds: it's a _stack_ of data. It functions similarly to how you would imagine a stack of anything would behave. For example, say you have a stack of papers each with its own information on it. You can only see once on top of the stack (peek), you add more to the stack by placing papers one at a time on top (push), and you grab papers to use one at a time from the top (pop). Stacks use the 'last in, first out' (LIFO) model for processing data, meaning the last data that was put on the stack is the first to come out.

## API & Efficiency

Stacks have three main functions:

* push: this allows the user to add elements to the stack
* pop: this allows the user to remove and return elements on the stack
* peek: this allows the user to view the value on top of the stack without removing it

All of these function are O(1) since all of these function happen at one location (the top of the stack), therefore no traversing is needed and each operation can be completed in constant time (no looping required).

ref: https://www.programiz.com/dsa/stack

stack.png.webp

## Implementation

Stacks are usually implemented as an array or as a list where it can then use the builtin functions of those data structures to push/pop/peek.

```
# example of a stack implemmentation

class Stack:

    def __init__(self):
        self.values = []
   
    def is_empty(self):
        if self.values == []:
            return True
        else:
            return False

    def pop(self):
        if self.is_empty():
            raise(IndexError)
        else:
            return self.values.pop()

    def peek(self):
        if self.is_empty():
            raise(IndexError)
        else:
            return self.values[-1]
    
    def push(self, val):
        self.values.append(val)
```

## Use Cases

Stacks are useful for keeping track of data that is being used. For example (I believe) that every function implemented into any code uses a stack to keep track of parameters and return values. In the implementation above, is_empty is pushed the address of the object 'self' then pops the boolean value for 'true/false'. Internally, computers utilize stacks for storing all data at addresses pointed to by a 'stack pointer.' So in a sense, every program uses a stack (internally). Stacks are useful for anytime someone needs to back track through data in reverse order: one simple example could be implementing a stack so that you can write it in reverse, or maybe even rendering a mirror image of something. Also useful for navigating through websites or keeping track of search histories: all previous websites are stored on a stack for which you can then navigate back to.

* enqueue: this allows the user to add elements to the rear of the queue
* dequeue: this allows the user to remove and return elements from the front of the queue

All of these function are O(1) since all of these function happen at a stored location (rear and front), therefore no traversing is needed and each operation can be completed in constant time (no looping required).

## Implementation

Queues are usually implemented as a singly linked list, since this improves the efficiency of the push/pop functions. In linked lists pushing and popping happens at constant time O(1).

```
# O(1) because assigning vals and appending vals are both constant time
def enqueue(self, item):
    self.items += 1
    self.data.append(item)
    
# O(1) because is_empty, raising error, assign val, and popleft() are all constant time
def dequeue(self):
    if self.is_empty():
        raise(ValueError)
    else:
        self.items -= 1
        return self.data.popleft()
```

## Use Cases

Queue are helpful in any scenario when a system is recieving data that needs to be processed one at a time. This helps the data to be read in correct order and in an organized way. For example something as simple as typing on a keyboard uses a queue to recieve key inputs and then display them in order (this only really is helpful for fast typers but still).

# Binary Search Trees

## Concept

A binary search tree (BST) is a node-based, hierarchical tree data structure. Each node has a key along with pointers to left, right, and parent nodes. A tree starts with a root node where all nodes will branch off of. A BST is organized so that all nodes left of the parent node are less than its key value and all nodes to the right are greater than the parent key. The sections are divided into subtrees that all follow the same orginazation scheme.

## API & Efficiency

Typical operations for a BST include:
* search(key): finds the node containing the specified key. 
* insert(node): inserts a specified node to an appropiate location on the tree.
* delete(key): deletes the node with the specified key
* traversal operations will be discussed in a later section

All functions use a similar traversal method so all their time complexities are similar. On average case its O(logn) and on worst case its O(n). In an average case during any operation traversal allows to skip half the tree which is proportional to the logorithm of the number of items stored in the tree. On worst case the tree is unbalanced/skewed so that the program has to traverse through all nodes in the tree

```
# examples of insert and search methods pulled from my own binary search implementation

def insert(self, child):
    if self.left is None or self.right is None:
        if self.key >= child.key:
            self.left = child
        else: 
            self.right = child
        child.parent = self
    else:
        if child.key <= self.key:
            self.left.insert(child)
        else:
            self.right.insert(child)

def search(self, key):
    if self.key == key:
        return self
    elif self.is_leaf():
        return None
    elif key < self.key:
        return self.left.search(key)
    elif key > self.key:
        return self.right.search(key)
```

## Tree Traversal

There are three main tree traversal algorithms that all occcur in three main steps (but in different orders):

1.   Preorder: first visit root, then visit left subtree, last visit right subtree
2.   Inorder: first vist left subtree, then visit root, last visit right subtree
3.   Postorder: first vist left subtree, then visit right subtree, last visit root

In preorder, visiting occurs by starting at the root, moving down the left branch until reaching a leaf, where it will recurse out by one node, move right one, then left until reaching another node and then the recursion continues untill all nodes have been visited.

In inorder, visiting occurs by starting at the left most node, recursing out one node, moving right one, then finding the left most leaf from that node, then recursing through that until the root is reached, then repeating the same process for the right subtree.

In post order, the same steps are followed except the root is now visited last.

# Priority Queue

##Concept

Like the name suggests, priority queues are function just like queue, but they are organized by an items priority. Priority is simple assigned by a number attached to and item being inserted into the priority queue. Once inserted the data structure inserts the item at the correct location in the priority queue. Highest priority is always at the end of the queue and this is where items get dequeued.

ref: https://www.programiz.com/dsa/priority-queue

Introduction.png.webp

##API & Efficiency

Like a queue, a priority queue has two main functions:

* enqueue: the object is inserted into the queue and moved to its correct location based on its priority value
* dequeue: the object with highest priority is removed from the queue and returned

When implemented as a max/min binary heap, which is the most common approach, the time efficiency for both functions is O(logn) or O(h) for worst and average case since inserting and deleting require traversing the height of the tree.

ref: https://algs4.cs.princeton.edu/24pq/

heap-ops.png

##Implementation

As previously stated, priority queues are most commonly implemented as min/max binary heaps, which are a type od binary trees, and abstracted even further are just arrays. In a binary heap the top value (root) is either the highest (max heap) or lowest (min heap) value in the tree and all values below the root have to be less than (for max heaps) or greater than (for min heaps). 

If the root is at the 0th index then all child indices are calculated using: child_i = 2 * parent_i + 1 for left and child_i = 2 * parent_i + 1 for right
And the parent indices are calculated from: parent_i = floor(child_i - 1)/2

For implementing insert functions, the value is first appended then it is sifted/percolated/bubbled/heaped up until it reaches its correct location in the heap order
Implementing delete functions, however, the root node (highest priority by value) is popped and then is swapped with the last value in the list where it is then sifted down to it's correct position. 1) that's why insert and delete is O(logn) and 2) that is how heap order is maintained in the process.

Alternatively, a priority queue can be implemented as a linked list, however this worsens the time complexity of inserts to O(n) since on every insert a linear sort has to occur, but it does improve the delete since it just pops the last/first node.

```
# example of enqueue and dequeue for priority queue using max heap
# (taken from my max heap implementation)

# O(h) worst case must sift up hieght of heap 
def insert(self, val):
    self._data.append(val)
    self._sift_up(self._last_index())

# O()
def delete(self):
    if self._is_empty():
        return None
    elif self._size() == 1:  
        return self._data.pop(0)
    else:
        root = self._value_at(0)
        self._swap(0, -1)
        self._data.remove(root)
        self._sift_down(0)
        return root

# O(1) series of conditional checks and getting values of known indices
def _obeys_heap_property_at_index(self, index):
    curr = self._value_at(index)
    if self._has_left_child(index):
        if self._left_child(index) > curr:
            return False
        if self._has_right_child(index):
            if self._right_child(index) > curr:
                return False
    return True

# O(1) boolean comparison
def _parent_obeys_heap_order(self, index):
    curr = self._value_at(index)
    parent = self._parent(index)
    if index == 0:
        return True
    elif parent < curr:
        return False
    else:
        return True

# O(1) assigning indices happens in constant time
# ref: https://www.geeksforgeeks.org/python-program-to-swap-two-elements-in-a-list/ 
def _swap(self, a, b):
    data = self._data
    data[a], data[b] = data[b], data[a]

# O(h) worst case scenario you have to swap down the hieght of the heap
def _sift_down(self, index):
    if self._obeys_heap_property_at_index(index):
        return
    child_i = self._greater_child_index(index)
    self._swap(index, child_i)
    return self._sift_down(child_i)
    
# O(h) worst case scenario you have to swap up the hieght of the heap
def _sift_up(self, index):
    parent_i = self._parent_index(index)
    if self._obeys_heap_property_at_index(index) and self._parent_obeys_heap_order(index):
        return
    self._swap(parent_i, index)
    return self._sift_up(parent_i)
```

# AVL Trees

##Concept

AVL tree is named after its founders Adelson-Velskii and Landis. AVL trees are self-balancing binary search trees. They use the same implementation however they include another data member called a balance factor. This balance factor is what triggers balancing/rotation methods within the program. The balancing factor is calculated by calculating the heights of the left and right subtrees of the node and finding the difference of the two. A tree is considered balanced if the balance factor is -1, 0, or 1.

##Sorting Demonstration

Sorting occurs whenever a nodes balance factor is greater than 1 or less than -1. There are four cases which result in which rotation operations are performed by the program. 

* Left-left case: the left tree is unbalanced (balance factor > 1) and the root's left node is greater than the inserted node; as a result the root node performs a right rotation at its location

* Right-right case: the right tree is unbalanced (balance factor < -1) and the root's right node is less than the inserted node; as a result the root node performs a left rotation at its location

* Left-right case: the left tree is unbalanced and the root's left node is less than the inserted node; as a result the left node rotates left and the root node rotates right

* Right-left case: the right tree is unbalanced and the root's right node is greater than the inserted node; as a result the right node rotates right and the root rotates left

##Implementation

An AVL tree is implemented the same way a binary search tree is implemented: using a binary heap. There is extra functionality added to the insert method that handles the balancing/rotate methods based on the key values and balance factor. Because AVL trees will always be balanced the worst case scenario for inserting, deleting, and searching will always be O(logn) since there will never be a scewed AVL tree (unlike regular BSTs). Rotations also occur at O(1) since it simply reassigns values/pointers.

```
# example of insert operations and rotates taken from my AVL tree implementation 
# (rebalance method not included, simply finds the heights of each tree and calculates the difference)

    # O(logn) since you only ever traverse the height of the tree and rotations occur in constant time
    def insert(self, node_to_insert):
        if node_to_insert.key <= self.key:
            if not self._has_left_child():
                self.left = node_to_insert
                self.left._rebalance()
            else:
                self.left.insert(node_to_insert)
        else:
            if not self._has_right_child():
                self.right = node_to_insert
                self.right._rebalance()
            else:
                self.right.insert(node_to_insert)

        self._rebalance()                               # rebalance root
        self.height = self._max_of_child_heights() + 1  # increase height by max of children heights' plus one

        if self.balance_factor > 1 and node_to_insert.key < self.left.key:      # left left case
            self = self._right_rotate()
        elif self.balance_factor < -1 and node_to_insert.key > self.right.key:  # right right case
            self = self._left_rotate()
        elif self.balance_factor > 1 and node_to_insert.key > self.left.key:    # left right case
            self.left = self.left._left_rotate()
            self = self._right_rotate()
        elif self.balance_factor < -1 and node_to_insert.key < self.right.key:  # right left case
            self.right = self.right._right_rotate()
            self = self._left_rotate()
        return self                                     # return root
            
    # O(1) Right rotate around self, assigning values happens in constant time
    # ref: https://www.geeksforgeeks.org/insertion-in-an-avl-tree/ 
    def _right_rotate(self):
        y = self.left
        T3 = y.right 

        y.right = self
        self.left = T3

        self.height = self._max_of_child_heights() + 1
        y.height = y._max_of_child_heights() + 1

        self._rebalance()       
        y._rebalance()

        return y
    
    # O(1) Left rotate around self, since reassigning values occurs in constant time
    def _left_rotate(self):
        y = self.right
        T2 = y.left

        y.left = self
        self.right = T2

        self.height = self._max_of_child_heights() + 1
        y.height = y._max_of_child_heights() + 1

        self._rebalance()       
        y._rebalance()

        return y
```
